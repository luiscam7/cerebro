```python
import json
import matplotlib.pyplot as plt
```

```python
file = 'test.json'

with open(file, 'r') as f:
    data = json.load(f)
```

```python
data.keys()
```

```python
data['powerline_noise_detected']
```

```python
data['ecg_noise_detected']
```

```python
psd = data['magnitude_spectral_density']
freqs = psd['freq']

for chann in psd.keys():

    plt.title('Power Spectral Density')
    plt.plot(freqs, psd[chann], label= f"{chann}")
    plt.ylabel('Power (uV**2/Hz)')
    plt.xlabel('Frequency (Hz)')
    plt.legend()
    plt.show()
```

```python
psd = data['regional_power_spectral_density']
freqs = psd['freqs']

regions = ['frontal', 'central', 'posterior']

for region in regions:
    plt.title('Power Spectral Density')
    plt.plot(freqs, psd[region], label= f"{region}")
    plt.ylabel('Power (uV**2/Hz)')
    plt.xlabel('Frequency (Hz)')
    plt.xlim(3,20)
    plt.legend()
plt.show()
```
